```python
import numpy as np
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from nltk.corpus import stopwords
```

```python
meta_raw = "Meta ranks among the largest American information technology companies, alongside other Big Five corporations Alphabet (Google), Amazon, Apple, and Microsoft. The company was ranked #31 on the Forbes Global 2000 ranking in 2023. In 2022, Meta was the company with the third-highest expenditure on research and development worldwide, with R&D expenditure amounting to US$35.3 billion. Meta has also acquired Oculus (which it has integrated into Reality Labs), Mapillary, CTRL-Labs, and a 9.99% stake in Jio Platforms; the company additionally endeavored into non-VR hardware, such as the discontinued Meta Portal smart displays line and partners with Luxottica through the Ray-Ban Stories series of smartglasses."
```

```python
meta_blob = TextBlob(meta_raw)
meta_text = meta_blob.words
```

```python
print(meta_text)
```

```python
meta_text_str = ' '.join(meta_text)
wc = WordCloud().generate(meta_text_str)
plt.axis("off")
plt.imshow(wc)
```

```python
wc = WordCloud(background_color='white', colormap = 'binary',
 	stopwords = ['meta'], width = 800, height = 500).generate(meta_text_str)
plt.axis("off")
plt.imshow(wc)
```
